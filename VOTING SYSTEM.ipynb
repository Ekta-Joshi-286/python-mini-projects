```python
# Voting System

# Step 1: Store candidates and votes in a dictionary
candidates = {}
voters = set()

# Step 2: Function to register candidates
def register_candidates():
    n = int(input("Enter number of candidates: "))
    for i in range(n):
        name = input(f"Enter candidate {i+1} name: ")
        candidates[name] = 0
    print("‚úÖ Candidates registered successfully!\n")

# Step 3: Function for voting
def voting():
    while True:
        voter_id = input("Enter your voter ID (or 'exit' to stop voting): ")
        if voter_id.lower() == "exit":
            break
        if voter_id in voters:
            print("‚ö†Ô∏è You have already voted!")
        else:
            print("\nCandidates:")
            for name in candidates:
                print("-", name)
            choice = input("Enter candidate name to vote for: ")
            if choice in candidates:
                candidates[choice] += 1
                voters.add(voter_id)
                print("‚úÖ Vote recorded!\n")
            else:
                print("‚ùå Invalid candidate name.\n")

# Step 4: Function to display results
def show_results():
    print("\n------ ELECTION RESULTS ------")
    for name, votes in candidates.items():
        print(f"{name}: {votes} votes")
    winner = max(candidates, key=candidates.get)
    print(f"\nüèÜ Winner: {winner} with {candidates[winner]} votes!")

# Step 5: Run the voting system
print("üéØ Welcome to the Voting System!\n")
register_candidates()
voting()
show_results()
```
