```python
# To-Do List Application

# Step 1: Create an empty list to store tasks
tasks = []

# Step 2: Define a simple menu
def show_menu():
    print("\n------ TO-DO LIST MENU ------")
    print("1. Add Task")
    print("2. View Tasks")
    print("3. Mark Task as Done")
    print("4. Delete Task")
    print("5. Exit")

# Step 3: Run program in a loop
while True:
    show_menu()
    choice = input("Enter your choice (1-5): ")

    # Add Task
    if choice == "1":
        task_name = input("Enter task name: ")
        tasks.append({"task": task_name, "done": False})
        print("‚úÖ Task added successfully!")

    # View Tasks
    elif choice == "2":
        if not tasks:
            print("No tasks found.")
        else:
            print("\n------ YOUR TASKS ------")
            for i, t in enumerate(tasks, start=1):
                status = "‚úÖ Done" if t["done"] else "‚ùå Not Done"
                print(f"{i}. {t['task']} - {status}")

    # Mark Task as Done
    elif choice == "3":
        if not tasks:
            print("No tasks to mark.")
        else:
            num = int(input("Enter task number to mark as done: "))
            if 1 <= num <= len(tasks):
                tasks[num - 1]["done"] = True
                print("üéØ Task marked as done!")
            else:
                print("Invalid task number.")

    # Delete Task
    elif choice == "4":
        if not tasks:
            print("No tasks to delete.")
        else:
            num = int(input("Enter task number to delete: "))
            if 1 <= num <= len(tasks):
                removed = tasks.pop(num - 1)
                print(f"üóëÔ∏è Deleted task: {removed['task']}")
            else:
                print("Invalid task number.")

    # Exit
    elif choice == "5":
        print("üëã Exiting To-Do List. Have a nice day!")
        break

    else:
        print("Invalid choice. Please try again.")
```
