```python
# Simple Banking Ledger

ledger = []  # List to store all transactions
balance = 0  # Initial balance

def add_transaction():
    global balance
    date = input("Enter date (DD-MM-YYYY): ")
    t_type = input("Enter type (Deposit/Withdraw): ").title()
    amount = float(input("Enter amount: "))

    if t_type == "Deposit":
        balance_change = amount
    elif t_type == "Withdraw":
        balance_change = -amount
    else:
        print("Invalid transaction type!")
        return

    global ledger
    ledger.append({'date': date, 'type': t_type, 'amount': amount})
    print("âœ… Transaction recorded successfully!")

def view_history():
    if not ledger:
        print("\nNo transactions yet!")
    else:
        print("\n------ Transaction History ------")
        for t in ledger:
            print(f"Date: {t['date']}, Type: {t['type']}, Amount: â‚¹{t['amount']}")
        print("---------------------------------")

def show_balance():
    total = 0
    for t in ledger:
        if t['type'] == "Deposit":
            total += t['amount']
        elif t['type'] == "Withdraw":
            total -= t['amount']
    print(f"\nðŸ’µ Current Balance: â‚¹{total}")

while True:
    print("\n------ BANKING MENU ------")
    print("1. Add Transaction")
    print("2. View Transaction History")
    print("3. Show Balance")
    print("4. Exit")

    choice = input("Enter your choice (1â€“4): ")

    if choice == '1':
        add_transaction()
    elif choice == '2':
        view_history()
    elif choice == '3':
        show_balance()
    elif choice == '4':
        print("ðŸ‘‹ Thank you for using the Banking Ledger!")
        break
    else:
        print("Invalid choice! Please try again.")
```
