```python
# Shopping Cart System

# Step 1: List of available items and prices
items = {
    "Apple": 50,
    "Banana": 20,
    "Milk": 60,
    "Bread": 40,
    "Eggs": 70
}

cart = []  # Empty list to store items added to the cart

while True:
    print("\n--------- SHOPPING MENU ---------")
    print("1. Show Items")
    print("2. Add Item to Cart")
    print("3. Remove Item from Cart")
    print("4. View Cart")
    print("5. Checkout and Exit")

    choice = input("Enter your choice (1-5): ")

    if choice == '1':
        print("\nAvailable Items:")
        for name, price in items.items():
            print(f"{name}: ₹{price}")

    elif choice == '2':
        item = input("Enter item name to add: ").title()
        if item in items:
            cart.append(item)
            print(f"{item} added to cart.")
        else:
            print("Item not found!")

    elif choice == '3':
        item = input("Enter item name to remove: ").title()
        if item in cart:
            cart.remove(item)
            print(f"{item} removed from cart.")
        else:
            print("Item not in cart!")

    elif choice == '4':
        if not cart:
            print("\nYour cart is empty!")
        else:
            print("\nItems in your cart:")
            for i in cart:
                print(f"- {i} (₹{items[i]})")

    elif choice == '5':
        if not cart:
            print("Your cart is empty! Goodbye!")
            break
        total = sum(items[i] for i in cart)
        print(f"\nTotal before discount: ₹{total}")

        # Simple discount logic
        if total >= 500:
            discount = total * 0.10  # 10% discount
        elif total >= 200:
            discount = total * 0.05  # 5% discount
        else:
            discount = 0

        final_total = total - discount
        print(f"Discount: ₹{discount:.2f}")
        print(f"Final Bill Amount: ₹{final_total:.2f}")
        print("Thank you for shopping with us!")
        break

    else:
        print("Invalid choice! Please try again.")
```
