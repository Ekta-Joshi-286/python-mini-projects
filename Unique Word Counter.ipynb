```python
# Unique Word Counter

import string  # To remove punctuation

# Step 1: Take paragraph input
text = input("Enter a paragraph:\n")

# Step 2: Clean text - remove punctuation & convert to lowercase
for p in string.punctuation:
    text = text.replace(p, "")
text = text.lower()

# Step 3: Split text into words
words = text.split()

# Step 4: Use set to find unique words
unique_words = set(words)
print(f"\nTotal unique words: {len(unique_words)}")

# Step 5: Count frequency of each word
word_count = {}
for w in words:
    word_count[w] = word_count.get(w, 0) + 1

# Step 6: Sort and display top 5 frequent words
sorted_words = sorted(word_count.items(), key=lambda x: x[1], reverse=True)

print("\n------ TOP 5 MOST FREQUENT WORDS ------")
for word, count in sorted_words[:5]:
    print(f"{word}: {count} times")
```
